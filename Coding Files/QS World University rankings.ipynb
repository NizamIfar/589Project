```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

### Extract the 2024 rankings from World rankings
##### Focus on the attributes required for the analysis

```python

Y2024_input_file = 'C:/Users/sahit/Desktop/MSDS/Advnaced_Python/Project/2024_QS_World_University_Rankings.xlsx'

Y2024_df = pd.read_excel(Y2024_input_file,header=[1, 2, 3])

Y2024_df = Y2024_df.dropna()
Y2024_df = Y2024_df.drop(columns=['Sustainability'])
Y2024_df = Y2024_df[Y2024_df[('Location', 'Location','location')] == 'United States']
Y2024_df = Y2024_df[~ (Y2024_df[(2024,'RANK','rank display')].apply(lambda x: ('-' in str(x)) or ('+' in str(x))))]
Y2024_df[(2024,'RANK','rank display')] = Y2024_df[(2024,'RANK','rank display')].apply(lambda x: str(x).replace('=',''))
Y2024_df = Y2024_df[~ (Y2024_df[(2023,'RANK','rank display2')].apply(lambda x: ('-' in str(x)) or ('+' in str(x))))]
Y2024_df[(2023,'RANK','rank display2')] = Y2024_df[(2023,'RANK','rank display2')].apply(lambda x: str(x).replace('=',''))
Y2024_df[('Overall','SCORE','Overall Score')] = Y2024_df[('Overall','SCORE','Overall Score')].apply(lambda x: str(x).replace('-',''))
```

### Extract the 2023 rankings from World rankings
##### Focus on the attributes required for the analysis

```python
# Replace 'file.xlsx' with the path to your Excel file
Y2023_input_file = 'C:/Users/sahit/Desktop/MSDS/Advnaced_Python/Project/2023_QS_World_University_Rankings.xlsx'

Y2023_df = pd.read_excel(Y2023_input_file,header=[1, 2, 3])

Y2023_df = Y2023_df.dropna()
Y2023_df.drop(('Classification', 'AGE'), axis=1, inplace=True)
Y2023_df = Y2023_df[Y2023_df[('Location', 'Location','location')] == 'United States']
Y2023_df = Y2023_df[~ (Y2023_df[(2023,'RANK','rank display')].apply(lambda x: ('-' in str(x)) or ('+' in str(x))))]
Y2023_df[(2023,'RANK','rank display')] = Y2023_df[(2023,'RANK','rank display')].apply(lambda x: str(x).replace('=',''))
Y2023_df = Y2023_df[~ (Y2023_df[(2022,'RANK','rank display2')].apply(lambda x: ('-' in str(x)) or ('+' in str(x))))]
Y2023_df[(2022,'RANK','rank display')] = Y2023_df[(2022,'RANK','rank display2')].apply(lambda x: str(x).replace('=',''))
Y2023_df[('Academic Reputation','RANK','ar rank')] = Y2023_df[(('Academic Reputation','RANK','ar rank'))].apply(lambda x: str(x).replace('+',''))
Y2023_df[('Employer Reputation','RANK','er rank')] = Y2023_df[(('Employer Reputation','RANK','er rank'))].apply(lambda x: str(x).replace('+',''))
Y2023_df[('Employment Outcomes','RANK','ger rank')] = Y2023_df[(('Employment Outcomes','RANK','ger rank'))].apply(lambda x: str(x).replace('+',''))
```

### Extract the 2017 to 2022 rankings from World rankings
##### Focus on the attributes required for the analysis

```python
# Replace 'file.xlsx' with the path to your Excel file
Pre_Y2023_input_file = 'C:/Users/sahit/Desktop/MSDS/Advnaced_Python/Project/2017-to-2022 QS-world-university-ranking.csv'

# Read the Excel file into a DataFramedf = df.reset_index(drop=True)
Pre_Y2023_df = pd.read_csv(Pre_Y2023_input_file)

Pre_Y2023_df = Pre_Y2023_df.dropna()

Pre_Y2023_df = Pre_Y2023_df[Pre_Y2023_df['country'] == 'United States']
Pre_Y2023_df = Pre_Y2023_df[~ (Pre_Y2023_df['rank_display'].apply(lambda x: ('-' in str(x)) or ('+' in str(x))))]
# print(Y2024_df[(2023,'RANK','rank display2')])
```

### Analyze 2024 rankings

```python
# Y2024_df_1 = Y2024_df[(2024,'RANK','rank display')].copy()
Y2024_df_1 = pd.DataFrame()
Y2024_df_1['Rank 2024'] = pd.DataFrame(Y2024_df[(2024,'RANK','rank display')].tolist())
Y2024_df_1['Institution'] = pd.DataFrame(Y2024_df[('Institution Name','Unnamed: 2_level_1','institution')].tolist())

Y2017_df_1 = pd.DataFrame()
selected_year = Pre_Y2023_df[Pre_Y2023_df['year'] == 2017]
Y2017_df_1['Rank 2017']   = pd.DataFrame(selected_year['rank_display'].tolist())
Y2017_df_1['Institution'] = pd.DataFrame(selected_year.iloc[:,0].tolist())
Y2017_df_1['Overall'] = pd.DataFrame(Y2024_df[('Overall','SCORE','Overall Score')].tolist())

Y2018_df_1 = pd.DataFrame()
selected_year = Pre_Y2023_df[Pre_Y2023_df['year'] == 2018]
Y2018_df_1['Rank 2018']   = pd.DataFrame(selected_year['rank_display'].tolist())
Y2018_df_1['Institution'] = pd.DataFrame(selected_year.iloc[:,0].tolist())

Y2019_df_1 = pd.DataFrame()
selected_year = Pre_Y2023_df[Pre_Y2023_df['year'] == 2019]
Y2019_df_1['Rank 2019']   = pd.DataFrame(selected_year['rank_display'].tolist())
Y2019_df_1['Institution'] = pd.DataFrame(selected_year.iloc[:,0].tolist())

Y2020_df_1 = pd.DataFrame()
selected_year = Pre_Y2023_df[Pre_Y2023_df['year'] == 2020]
Y2020_df_1['Rank 2020']   = pd.DataFrame(selected_year['rank_display'].tolist())
Y2020_df_1['Institution'] = pd.DataFrame(selected_year.iloc[:,0].tolist())

Y2021_df_1 = pd.DataFrame()
selected_year = Pre_Y2023_df[Pre_Y2023_df['year'] == 2021]
Y2021_df_1['Rank 2021']   = pd.DataFrame(selected_year['rank_display'].tolist())
Y2021_df_1['Institution'] = pd.DataFrame(selected_year.iloc[:,0].tolist())

Y2022_df_1 = pd.DataFrame()
selected_year = Pre_Y2023_df[Pre_Y2023_df['year'] == 2022]
Y2022_df_1['Rank 2022']   = pd.DataFrame(selected_year['rank_display'].tolist())
Y2022_df_1['Institution'] = pd.DataFrame(selected_year.iloc[:,0].tolist())

Y2023_df_1 = pd.DataFrame()
Y2023_df_1['Rank 2023'] = pd.DataFrame(Y2023_df[(2023,'RANK','rank display')].tolist())
Y2023_df_1['Institution'] = pd.DataFrame(Y2023_df[('Institution Name','Unnamed: 2_level_1','institution')].tolist())

merged_df = pd.merge(Y2017_df_1, Y2018_df_1, on=('Institution'), how='outer')
merged_df = merged_df.dropna()
merged_df = pd.merge(merged_df, Y2019_df_1, on=('Institution'), how='outer')
merged_df = merged_df.dropna()
merged_df = pd.merge(merged_df, Y2020_df_1, on=('Institution'), how='outer')
merged_df = merged_df.dropna()
merged_df = pd.merge(merged_df, Y2021_df_1, on=('Institution'), how='outer')
merged_df = merged_df.dropna()
merged_df = pd.merge(merged_df, Y2022_df_1, on=('Institution'), how='outer')
merged_df = merged_df.dropna()
merged_df = pd.merge(merged_df, Y2023_df_1, on=('Institution'), how='outer')
merged_df = merged_df.dropna()
merged_df = pd.merge(merged_df, Y2024_df_1, on=('Institution'), how='outer')
merged_df = merged_df.dropna()



merged_df['Rank 2024']=merged_df['Rank 2024'].astype(int)
merged_df['Rank 2023']=merged_df['Rank 2023'].astype(int)
merged_df['Rank 2022']=merged_df['Rank 2022'].astype(int)
merged_df['Rank 2021']=merged_df['Rank 2021'].astype(int)
merged_df['Rank 2020']=merged_df['Rank 2020'].astype(int)
merged_df['Rank 2019']=merged_df['Rank 2019'].astype(int)
merged_df['Rank 2018']=merged_df['Rank 2018'].astype(int)
merged_df['Rank 2017']=merged_df['Rank 2017'].astype(int)


# print(Y2024_df_1.head())
# print(Y2023_df_1.head())
# print(Y2022_df_1.head())
print(merged_df)
```

### Correlation between all the scores

```python
corr_columns = Y2024_df_1.columns.tolist()
corr_columns.remove('Institution')
correlation_matrix = Y2024_df_1[corr_columns].corr()
# print(correlation_matrix)

# print(correlation_matrix)
# Plot the correlation matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

### Correlation Between World University Rank and Employment Outcomes Rank
