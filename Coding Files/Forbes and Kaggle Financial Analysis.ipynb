```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# The API endpoint from which to fetch the table data
api_url = 'https://www.forbes.com/forbesapi/org/top-colleges/2023/position/true.json?limit=1000&fields=organizationName,academics,state,financialAid,rank,medianBaseSalary,campusSetting,studentPopulation,squareImage,uri,description,grade'

# Make a GET request to the API endpoint
response = requests.get(api_url)
print(response)
json_data = response.json()
# print(response.json())
```

```python
# Create a list to hold all the extracted information
extracted_info = []

# Loop through each organization in the organizationsLists
for org in json_data['organizationList']['organizationsLists']:
    # Extract the desired keys
    rank = org.get('rank')
    name = org.get('organizationName')
    state = org.get('state')
    org_type = org.get('academics', {}).get('type')
    
    # Initialize default values for grant aid and debt
    avg_grant_aid = None
    avg_debt = None
    
    # Check if 'financialAid' key exists
    if 'financialAid' in org:
        # Use a generator expression to find the average grant aid and debt values
        avg_grant_aid = next((item['value'] for item in org['financialAid'].get('avgGrantAidByType', []) if item['key'] == 'anyGrant'), None)
        avg_debt = next((item['value'] for item in org['financialAid'].get('avgLoansByType', []) if item['key'] == 'anyLoan'), None)
    
    median_salary = org.get('medianBaseSalary')
    financial_grade = org.get('grade')
    
    # Append a dictionary with the extracted data to the list
    extracted_info.append({
        'RANK': rank,
        'NAME': name,
        'STATE': state,
        'TYPE': org_type,
        'AV. GRANT AID': avg_grant_aid,
        'AV. DEBT': avg_debt,
        'MEDIAN 10-YEAR SALARY': median_salary,
        'FINANCIAL GRADE': financial_grade
    })

# For demonstration, print the extracted information
if extracted_info:
    print(extracted_info[0])
else:
    print("The list is empty.")
```

```python
# Format the currency fields as strings with a dollar sign, commas for thousands, and no decimal places
for item in extracted_info:
    if item['AV. GRANT AID'] is not None:
        item['AV. GRANT AID'] = f"${item['AV. GRANT AID']:,.0f}"  # No decimal points
    else:
        item['AV. GRANT AID'] = None

    if item['AV. DEBT'] is not None:
        item['AV. DEBT'] = f"${item['AV. DEBT']:,.0f}"  # No decimal points
    else:
        item['AV. DEBT'] = None

    if item.get('MEDIAN 10-YEAR SALARY') is not None:
        item['MEDIAN 10-YEAR SALARY'] = f"${item['MEDIAN 10-YEAR SALARY']:,.0f}"  # No decimal points
    else:
        item['MEDIAN 10-YEAR SALARY'] = None

# Now create the DataFrame
df = pd.DataFrame(extracted_info)
#chatGPT: how to print first 2 rows of df
print(df.head(1))
# When saving to an Excel file, the formatted strings will be preserved
df.to_excel('2023-forbes-college-rankings.xlsx', index=False)
print(df.head(1))
```
