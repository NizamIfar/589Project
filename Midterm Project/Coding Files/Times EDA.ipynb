```python
import pandas as pd

file_path = '/Users/mazinrafi/Downloads/Times Rankings 2024 - Sheet1.csv'
times_ranking_data = pd.read_csv(file_path)
times_ranking_data.head()

#many 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
times_ranking_data['Overall Score'] = pd.to_numeric(times_ranking_data['Overall Score'], errors='coerce')

# Plotting the distribution of overall scores
plt.figure(figsize=(10, 6))
sns.histplot(times_ranking_data['Overall Score'], bins=20, kde=False, color='blue')
plt.title('Distribution of Overall Scores')
plt.xlabel('Overall Score')
plt.ylabel('Number of Universities')
plt.grid(True)
plt.show()
```

```python
# Converting 'Industry Score' to numeric for visualization
times_ranking_data['Industry Score'] = pd.to_numeric(times_ranking_data['Industry Score'], errors='coerce')

# Ensuring 'Rank' is in the correct numeric format
times_ranking_data['Rank '] = pd.to_numeric(times_ranking_data['Rank '], errors='coerce')
# Re-plotting the relationship between the rank and the industry score, ensuring the 'Rank' column is used correctly
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Rank ', y='Industry Score', data=times_ranking_data)
plt.title('Rank vs. Industry Score')
plt.xlabel('Rank')
plt.ylabel('Industry Score')
plt.grid(True)
plt.tight_layout()  # Adjust layout to make sure everything fits neatly
plt.show()
```

```python
# Enhancing the plot for better aesthetics and readability
plt.figure(figsize=(12, 7))
sns.scatterplot(x='Rank ', y='Industry Score', data=times_ranking_data, color='dodgerblue', edgecolor='black', s=50)

# Improving the title and labels
plt.title('University Rank vs. Industry Score', fontsize=16)
plt.xlabel('University Rank', fontsize=14)
plt.ylabel('Industry Score', fontsize=14)

# Refining the grid and layout
plt.grid(True, which='major', linestyle='--', linewidth=0.5, color='grey')
plt.tight_layout()  # Adjust layout to make sure everything fits neatly

# Show the plot with enhancements
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.scatterplot(x='Rank ', y='No. of students per staff', data=times_ranking_data, color='green', edgecolor='black', s=50)
plt.title('University Rank vs. Student-to-Faculty Ratio', fontsize=16)
plt.xlabel('University Rank', fontsize=14)
plt.ylabel('Number of Students per Faculty', fontsize=14)
plt.grid(True, which='major', linestyle='--', linewidth=0.5, color='grey')
plt.tight_layout()
plt.show()
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'times_ranking_data' is your DataFrame and it's already loaded and preprocessed
top_5_universities = times_ranking_data.sort_values(by='Overall Score', ascending=False).head(5)

plt.figure(figsize=(10, 6))
barplot = sns.barplot(x='Name\nCountry/Region', y='Overall Score', data=top_5_universities, palette='viridis')

# Iterate over the bars and use the `text` function to add labels
for p in barplot.patches:
    barplot.text(p.get_x() + p.get_width() / 2., p.get_height(), '%d' % int(p.get_height()), 
                 ha='center', va='bottom')

plt.title('Top 5 Universities by Overall Score')
plt.xlabel('University')
plt.ylabel('Overall Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```

```python
# Preparing the data: converting the percentage of interdisciplinary science research to numeric values
times_ranking_data['% Interdiscip. Science Research'] = times_ranking_data['% Interdiscip. Science Research'].str.replace('%', '').astype(float) / 100

# Plotting the relationship between the percentage of interdisciplinary science research and the overall score
plt.figure(figsize=(10, 6))
sns.scatterplot(x='% Interdiscip. Science Research', y='Overall Score', data=times_ranking_data)
plt.title('Overall Score vs. Percentage of Interdisciplinary Science Research')
plt.xlabel('Percentage of Interdisciplinary Science Research')
plt.ylabel('Overall Score')
plt.grid(True)
plt.show()
```

```python
# Correcting the column names
times_ranking_data.rename(columns={'Rank ': 'Rank', 'Name\nCountry/Region': 'Name'}, inplace=True)

# Convert 'Rank' to numeric values for visualization
times_ranking_data['Rank'] = pd.to_numeric(times_ranking_data['Rank'], errors='coerce')

# Plotting the distribution of university rankings
plt.figure(figsize=(10, 6))
sns.histplot(times_ranking_data['Rank'], bins=30, kde=False, color='skyblue')
plt.title('Distribution of University Rankings')
plt.xlabel('Rank')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Convert 'Global Rank' and 'No. of students per staff' to numeric for visualization
times_ranking_data['Global Rank'] = pd.to_numeric(times_ranking_data['Global Rank'], errors='coerce')
times_ranking_data['No. of students per staff'] = pd.to_numeric(times_ranking_data['No. of students per staff'], errors='coerce')

# Plotting the relationship between the number of students per staff and the global rank
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Global Rank', y='No. of students per staff', data=times_ranking_data)
plt.title('Global Rank vs. Number of Students per Staff')
plt.xlabel('Global Rank')
plt.ylabel('Number of Students per Staff')
plt.grid(True)
plt.show()
```

```python
# Correcting the extraction and processing of the 'Female:Male Ratio' data
female_male_ratios = times_ranking_data['Female:Male Ratio'].str.extract(r'(\d+) : (\d+)')
times_ranking_data['Female Ratio'] = female_male_ratios[0].astype(float) / (female_male_ratios[0].astype(float) + female_male_ratios[1].astype(float))

# Verifying the extraction and conversion
times_ranking_data[['Female:Male Ratio', 'Female Ratio']].head()

# Plotting the distribution of female ratios across universities
plt.figure(figsize=(10, 6))
sns.histplot(times_ranking_data['Female Ratio'], bins=20, kde=False, color='purple')
plt.title('Distribution of Female Ratios Across Universities')
plt.xlabel('Female Ratio')
plt.ylabel('Number of Universities')
plt.grid(True)
plt.show()
```

```python
# Selecting relevant columns for correlation analysis
correlation_data = times_ranking_data[['Overall Score', 'Teaching Score', 'Research Environment Score', 'Research Quality Score', 'Industry Score', 'International Outlook Score']]

# Calculating the correlation matrix
correlation_matrix = correlation_data.corr()

# Plotting the heatmap for the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix of University Scores and Rankings')
plt.show()
```
